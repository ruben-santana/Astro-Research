```python
#Importing and notebook setup
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import optimize
```

```python
#Data import
data= pd.read_csv('C:/users/rscsa/Research/radvel-master/Research/HD217014/HD217014.csv')

#Naming Variables from data
t = np.array(data.time)
vel = np.array(data.mnvel)
errvel = np.array(data.errvel)

#best fit time vector
time = np.linspace(t[0]-1,t[-1]+1,num=100000)
```

```python
#using Lomb-Scargle Power (CORRECT)
#from gatspy
from gatspy.periodic import LombScargleFast
model = LombScargleFast().fit(t, vel)
period, power = model.periodogram_auto(nyquist_factor=50)

#Plotting
plt.figure
plt.plot(period,power)
plt.ylabel('Power')
plt.xlabel('Period')# days
plt.xlim(0,8)
```

```python
# set range and find period (Correct)
model.optimizer.period_range=(2, 8)
period = model.best_period
print("period = {0}".format(period))
```

```python
#http://scipy-cookbook.readthedocs.io/items/FittingData.html
```

```python

# Generate data points with noise
range_min = 12
range_max = 26
#x-axis data
data_x = t[range_min:range_max]

#y-axis Data
data_y = vel[range_min:range_max]


# Target function
fitfunc = lambda p, x: p[0]*np.cos(2*np.pi/p[1]*x+p[2]) + p[3]*x
#Distance to the target function
errfunc = lambda p, x, y: fitfunc(p, x) - y 
# Initial guess for the first set's parameters
p0 = [50., period, 0., 0.]
p1, success = optimize.leastsq(errfunc, p0[:], args=(data_x, data_y))
print 'Initial Guess: [Amplitude, Period, Phase Shift, Y-Shift]'
print 'Initial Guess: ', p0
print 'Optimized: ', p1

Residuals = data_y - fitfunc(p1,data_x)
p0_r = [10., period, 0., 0.]
p2, success2 = optimize.leastsq(errfunc, p0_r[:], args=(data_x, Residuals))

print Residuals

time = np.linspace(data_x.min(), data_x.max(), 100000)
plt.plot(data_x, data_y, "o", time, fitfunc(p1, time),"b-")

# Legend the plot
plt.title('HD217014')
plt.xlabel('Time [JD-2.44e6]')
plt.ylabel('Mean Velocity[m/s]')
plt.legend(('Data', 'Fit'))
plt.xlim(data_x[0]-5,data_x[-1]+5)

ax1 = plt.axes()
plt.show()

plt.plot(data_x, Residuals, "o", time, fitfunc(p2, time),"b-")
plt.xlim(data_x[0]-5,data_x[-1]+5)
plt.ylim(-max(data_y)-10, max(data_y)+10)
plt.title('Residuals')
plt.xlabel('Time [JD-2.44e6]')
plt.ylabel('Mean Velocity[m/s]')
plt.legend(('Residuals', 'Fit'))
ax2 = plt.axes()
```
