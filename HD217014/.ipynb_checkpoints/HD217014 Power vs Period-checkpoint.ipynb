```python
#Importing and notebook setup
%matplotlib inline

import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.optimize as optimization
```

```python
#Data import
data= pd.read_csv('C:/users/rscsa/Research/radvel-master/HD217014/HD217014.csv')

t = np.array(data.time)
vel = np.array(data.mnvel)
errvel = np.array(data.errvel)
yerr = sum(errvel)/len(errvel)

# print t
# print vel
# print errvel
# print yerr
```

```python
#best fit time vector
time = np.linspace(t[0]-1,t[-1]+1,num=10000)

# print time
```

```python
#Finding Least-Square Function
# time = np.linspace(t[0]-1,t[-1]+1,t[-1]/10000.)
def func(x, amp, freq, phaseshift):
    return amp*np.sin(freq*x + phaseshift)


param = optimization.curve_fit(func,t, vel)[0]
# print optimization.leastsq(func,args=(t, vel,0,0,1)

best_fit = func(time, param[0], param[1], param[2])
print 'Amplitude: ', param[0]
print 'Frequency: ', param[1]
print 'Phase Shift: ', param[2]

#Data Plotting
plt.figure()
plt.xlabel('Time (JD-2.44e6)')
plt.ylabel('Mean Velocity')
plt.errorbar(t,vel,yerr=errvel, fmt='o', ecolor='red')
plt.plot(time,best_fit,color='green')
plt.title('HD217014')
plt.xlim(t[0],t[0]+200)
```

```python
#using Lomb-Scargle Power (CORRECT)
#from gatspy
from gatspy.periodic import LombScargleFast
model = LombScargleFast().fit(t, vel)
periods, power1 = model.periodogram_auto(nyquist_factor=80)

#Plotting
plt.figure
plt.plot(periods,power1)
plt.ylabel('Power')
plt.xlabel('Period')# days
plt.xlim(0,8)
```

```python
# set range and find period (Correct)
model.optimizer.period_range=(2, 8)
period = model.best_period
print("period = {0}".format(period))
```
